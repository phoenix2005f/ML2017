##  train_X,train_y

```python
def flatten(df):
    res=[]
    for i in range(len(df)):
        res.extend(df.iloc[i])
    
    return res

def train(X,y,l_rate,repeat):
    w = np.zeros(len(X[0]))
    s_gra=0
    for i in range(repeat):
        hypo = np.dot(X,w)
        loss = hypo-y
        cost = np.sum(loss**2)/len(X)
        cost_a = np.sqrt(cost)
        gra = np.dot(X.T,loss)
        s_gra += gra**2
        ada = np.sqrt(s_gra)
        w = w-l_rate * gra/ada
        print("iteration:{}| Cost:{}".format(i,cost_a))

    return w
```

```python
X=[]
y=[]

for i in range(12):
    for j in range(471):
        X.append([])
        flat = flatten(data.iloc[471*i+j:471*i+j+9])
        X[471*i+j].extend(flat)
        y.append(data.iloc[471*i+j+9]["PM2.5"])
X = np.array(X)
y = np.array(y)        
```

```python
X = np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
```

```python
l_rate = 10
repeat = 10000
w=train(X,y,l_rate,repeat)
```

## save model

```python
np.save("model.npy",w)
```
